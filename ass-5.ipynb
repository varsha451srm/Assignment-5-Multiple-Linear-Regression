# Calculating VIF

```python
rsq_rd = smf.ols('RD_Spend~Administration+Marketing_Spend+State',data=profit).fit().rsquared  
vif_rd = 1/(1-rsq_rd)

rsq_ad = smf.ols('Administration~RD_Spend+Marketing_Spend+State',data=profit).fit().rsquared  
vif_ad = 1/(1-rsq_ad) 

rsq_mt = smf.ols('Marketing_Spend~RD_Spend+Administration+State',data=profit).fit().rsquared  
vif_mt = 1/(1-rsq_mt)

rsq_st = smf.ols('State~RD_Spend+Administration+Marketing_Spend',data=profit).fit().rsquared  
vif_st = 1/(1-rsq_st)

# Storing vif values in a data frame
d1 = {'Variables':['RD_Spend','Administration','Marketing_Spend','State'],'VIF':[vif_rd,vif_ad,vif_mt,vif_st]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

# Model Deletion Diagnostics

## Detecting Influencers/Outliers

## Cook’s Distance

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

### Stem plot for cook's distance and index of row

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(profit)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

# High influence points

```python
from statsmodels.graphics.regressionplots import influence_plot
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k = profit.shape[1]
n = profit.shape[0]
leverage_cutoff = 3*((k + 1)/n)
```
